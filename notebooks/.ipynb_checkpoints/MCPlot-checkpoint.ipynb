```python
class MCCheck(object):
    
    def __init__(self, paraGenClass):
        self._Spectrum = paraGenClass._Spectrum
        self._mode = paraGenClass._mode
        return None
    
    def _datGen(self, RNuSM, RNuNP, RWIMP, Uncs, b0, exposure, modeHypothesis, modeReal):
        for i in range(100):
            nuiList = np.array(list(map(lambda d: np.random.normal(loc=1.,scale=d), Uncs)))
            if np.all(nuiList>0):
                break
        if modeHypothesis=='SMvsDMNP':
            if modeReal=='H0':
                dat = nuiList@RNuSM + b0
            elif modeReal=='H1':
                dat = nuiList@RNuNP + RWIMP + b0
            else:
                raise ValueError('Error in MCCheck._datGen')
        elif modeHypothesis=='DMvsNP':
            if modeReal=='H0':
                dat = nuiList@RNuNP + b0
            elif modeReal=='H1':
                dat = nuiList@RNuSM + RWIMP + b0
            else:
                raise ValueError('Error in MCCheck._datGen')
        else:
            raise ValueError('Error in MCCheck._datGen')
        return np.array(list(map(lambda d: np.random.poisson(lam=d), dat*exposure)))/exposure
    
    def _statisticGen_DMvsNP(self, datObs, RNuSM, RNuDiff, RWIMP, Uncs, b0, exposure):
        n = np.shape(Uncs)[0]
        H0Gen = lambda x: globalFunc(datObs, RNuSM, RWIMP, 0., RNuDiff, 1., x, b0, Uncs, exposure)
        H1Gen = lambda x: globalFunc(datObs, RNuSM, RWIMP, 1., RNuDiff, 0., x, b0, Uncs, exposure)
        H2Gen = lambda x: globalFunc(datObs, RNuSM, RWIMP, x[0], RNuDiff, x[1], x[2:], b0, Uncs, exposure)
        
        boundsH0 = np.transpose([np.zeros(n)+1e-15, np.ones(n)*3.])
        boundsH2 = np.transpose([np.zeros(n+2)+1e-15, np.append([1e4,1e4],np.ones(n)*3.)])
        likeDeno = differential_evolution(H1Gen, boundsH0, tol=0, atol=0.01)
        likeNume = differential_evolution(H0Gen, boundsH0, tol=0, atol=0.01)
        likeX = differential_evolution(H2Gen, boundsH2, tol=0, atol=0.01)
        #print(likeDeno," ",likeNume," ",[datDM])
        return [2*(likeNume.fun-likeDeno.fun), 2*(likeNume.fun-likeX.fun),\
                2*(likeDeno.fun-likeX.fun), likeNume.x, likeDeno.x, likeX.x, datObs]
    
    def _statisticGen_SMvsDMNP(self, datObs, RNuSM, RNuDiff, RWIMP, Uncs, b0, exposure):
        n = np.shape(Uncs)[0]
        H0Gen = lambda x: globalFunc(datObs, RNuSM, RWIMP, 0., RNuDiff, 0., x, b0, Uncs, exposure)
        H1Gen = lambda x: globalFunc(datObs, RNuSM, RWIMP, 1., RNuDiff, 1., x, b0, Uncs, exposure)
        H2Gen = lambda x: globalFunc(datObs, RNuSM, RWIMP, x[0], RNuDiff, x[1], x[2:], b0, Uncs, exposure)
        
        boundsH0 = np.transpose([np.zeros(n)+1e-15, np.ones(n)*3.])
        likeDeno = differential_evolution(H1Gen, boundsH0, tol=0, atol=0.01)
        likeNume = differential_evolution(H0Gen, boundsH0, tol=0, atol=0.01)
        return [2*(likeNume.fun-likeDeno.fun), likeNume.x, likeDeno.x, datObs]
    
    def StatGen(self, mDM, sigma0, exposure, n, b0Level, modeReal):
        RNuSM, NuUncs, RNuNP = self._Spectrum._RNuSM, self._Spectrum._NuUnc, self._Spectrum._RNuNP
        RNuDiff = RNuNP-RNuSM
        RWIMP = self._Spectrum._RWIMP*10**(sigma0+45.)
        b0 = self._Spectrum._b0Flat*b0Level
        modeHypothesis = self._mode
        
        datObsList = np.array(list(map(lambda i: \
                self._datGen(RNuSM, RNuNP, RWIMP, NuUncs, b0, exposure, modeHypothesis, modeReal), range(n))))
        
        if modeHypothesis == 'SMvsDMNP':
            statisticGen = self._statisticGen_SMvsDMNP
        elif modeHypothesis == 'DMvsNP':
            statisticGen = self._statisticGen_DMvsNP
        else:
            raise TypeError('No such mode')
        
        # 20231211 To accelerate the minimum searching, memory assignment of Mat needs to be changed
        RNuSM = np.transpose(np.ascontiguousarray(np.transpose(RNuSM)))
        RNuDiff = np.transpose(np.ascontiguousarray(np.transpose(RNuDiff)))
        
        time_start = time.perf_counter()
        pool = multiprocessing.Pool(4)
        multiple_results = [pool.apply_async(\
                statisticGen, (datObsList[i], RNuSM, RNuDiff, RWIMP, NuUncs, b0, exposure))\
                                             for i in range(n)]
        pool.close()
        pool.join()
        Stat= [res.get() for res in multiple_results]
        time_end = time.perf_counter()
        print("Time costed: {0} s.".format(time_end-time_start))
        return Stat
    
def singleMassCheck(mDM, paraGenClass, FogGenClass, modelPara):
    exposureEndLog, dExpoLog, CL, b0Level = 8., 0.1, 0.9, 0.
    paraGenClass.setModel(modelPara)
    paraGenClass.setDMMass(mDM)
    
    exposureInitLog=np.log10(0.1/np.sum(paraGenClass._Spectrum._RNuSM))
    res1 = FogGenClass.sigmaExpoListDMDiscoveryGen(paraGenClass, exposureEndLog, dExpoLog, CL, b0Level,\
                                                   exposureInitLog=exposureInitLog)
    if not isinstance(res1,list):
        return 1
    res2 = np.array([np.array([myFindRoot(d) for d in res1[2]]),res1[0]])
    x,y=res2[0], res2[1]
    fig, ax = plt.subplots(1, 3, figsize=(15,5))
    ax[0].plot(x, y, linestyle='-', linewidth=2)
    lfs = 20
    ax[0].set_xlabel(r"Exposure",fontsize=lfs)
    ax[0].set_ylabel(r'Cross section [cm$^{-2}$]',fontsize=lfs)
    ax[0].set_xlim([x[0],x[-1]])
    ax[0].tick_params(which='major',direction='in',width=2,length=5,pad=7)
    ax[0].tick_params(which='minor',direction='in',width=1,length=5)
    
    res3 = vsCLGen(paraGenClass, res2[1], res2[0], 0.)
    x,y=res3
    ax[1].plot(x, y, linestyle='-', linewidth=2)
    ax[1].set_xlabel(r"Exposure",fontsize=lfs)
    ax[1].set_ylabel(r'CL',fontsize=lfs)
    ax[1].set_xlim([x[0],x[-1]])
    DistinctionExposure, transitionIndices = vsExpoGen(mDM, res3, CL)
    if len(DistinctionExposure)>0:
        DistinctionSigma = np.interp(DistinctionExposure,res2[0],res2[1])
    else:
        DistinctionSigma = [-60., -60.]
    ax[1].plot([DistinctionExposure[0],DistinctionExposure[0]], [0,1], linestyle='--', color='r')
    ax[1].plot([DistinctionExposure[1],DistinctionExposure[1]], [0,1], linestyle='--', color='r')
    ax[1].plot(x, np.ones_like(x)*CL, linestyle='-.', color='purple')
    
    x = res2[1][:len(y)]
    ax[2].plot(x, y, linestyle='-', linewidth=2)
    ax[2].set_xlabel(r"Discovery Cross Section",fontsize=lfs)
    ax[2].set_ylabel(r'CL',fontsize=lfs)
    ax[2].set_xlim([x[0],x[-1]])
    
    ax[2].plot([DistinctionSigma[0],DistinctionSigma[0]], [0,1], linestyle='--', color='r')
    ax[2].plot([DistinctionSigma[1],DistinctionSigma[1]], [0,1], linestyle='--', color='r')
    ax[2].plot(x, np.ones_like(x)*CL, linestyle='-.', color='purple')
    print([DistinctionExposure, DistinctionSigma])
    print([10**np.array(DistinctionExposure), 10**np.array(DistinctionSigma)])
    return 

def myMCCheck(paraGenClass, MCDat, sigma0, exposure,nBins,xrange,ymax):
    paraGenClass.vsTempMatSet(sigma0, b0Level)
    vsParaGenFunc = paraGenClass.vsParaGen
    AsimovGenFunc = paraGenClass.vsAsimovGen
    thetaPrimeList = paraGenClass._thetaPrimeList
    mode = paraGenClass._mode
    
    temp = np.array([vsParaGenFunc(exposure, i) for i in range(2)])
    temp2 = np.array([AsimovGenFunc(sigma0, exposure,  b0Level, the) \
                      for the in thetaPrimeList])
    muList = [temp[0,0], temp[1,0], temp2[0], temp2[1]]
    sigList = [temp[0,1], temp[1,1], temp[0,1], temp[1,1]]
    
    fig, ax = plt.subplots(1, 1, figsize=(10,10))
    
    if mode == 'SMvsDMNP':
        StatList = np.transpose([[d[0]] for d in MCDat[0]])
    else:
        StatList = np.transpose([[d[0],d[1],d[2]] for d in MCDat[0]])
    comPlotDat = StatList[0]
    
    if mode == 'SMvsDMNP':
        StatList = np.transpose([[d[0]] for d in MCDat[1]])
    else:
        StatList = np.transpose([[d[0],d[1],d[2]] for d in MCDat[1]])
    comPlotDat = np.array([comPlotDat, StatList[0]])
    
    #GaussianPlot(ax[0][2], comPlotDat, nBins, muList, sigList,xrange)
    sigList = [temp[1,1], temp[0,1], temp[1,1], temp[0,1]]
    GaussianPlot(ax, comPlotDat, nBins, muList, sigList,xrange,ymax)
    
    return 

def myMCCheck2(paraGenClass, MCDat, sigma0, exposure,nBins,xrange,ymax,ax):
    paraGenClass.vsTempMatSet(sigma0, b0Level)
    vsParaGenFunc = paraGenClass.vsParaGen
    AsimovGenFunc = paraGenClass.vsAsimovGen
    thetaPrimeList = paraGenClass._thetaPrimeList
    mode = paraGenClass._mode
    
    temp = np.array([vsParaGenFunc(exposure, i) for i in range(2)])
    temp2 = np.array([AsimovGenFunc(sigma0, exposure,  b0Level, the) \
                      for the in thetaPrimeList])
    muList = [temp[0,0], temp[1,0], temp2[0], temp2[1]]
    sigList = [temp[0,1], temp[1,1], temp[0,1], temp[1,1]]
    
    if mode == 'SMvsDMNP':
        StatList = np.transpose([[d[0]] for d in MCDat[0]])
    else:
        StatList = np.transpose([[d[0],d[1],d[2]] for d in MCDat[0]])
    comPlotDat = StatList[0]
    
    if mode == 'SMvsDMNP':
        StatList = np.transpose([[d[0]] for d in MCDat[1]])
    else:
        StatList = np.transpose([[d[0],d[1],d[2]] for d in MCDat[1]])
    comPlotDat = np.array([comPlotDat, StatList[0]])
    
    #GaussianPlot(ax[0][2], comPlotDat, nBins, muList, sigList,xrange)
    sigList = [temp[1,1], temp[0,1], temp[1,1], temp[0,1]]
    GaussianPlot(ax, comPlotDat, nBins, muList, sigList,xrange,ymax)
    
    return 

def GaussianPlot(ax, StatList, nBins, mu, sig, xrange, ymax):
    print('Minimum: ', list(map(np.min, StatList)), ' Maximum: ', list(map(np.max, StatList)))
    xMin, xMax = np.floor(np.min(np.concatenate(StatList))), np.ceil(np.max(np.concatenate(StatList)))
    StatX = np.linspace(xMin,xMax,nBins)
    DatX = np.array(list(map(lambda x: [x,x], StatX[1:-1]))).flatten()
    DatX = np.append([StatX[0]], np.append(DatX, [StatX[-1]]))
    def datYGen(dat):
        StatY = np.histogram(dat, bins=StatX)[0]
        StatY = 1/(np.sum(StatY)*(StatX[1]-StatX[0]))*StatY
        DatY = np.array(list(map(lambda x: [x,x], StatY))).flatten()
        return DatY
    datYList = [datYGen(d) for d in StatList]
    xMin, xMax = np.min([DatX[np.where(d>1e-4)[0][0]] for d in datYList]), \
            np.max([DatX[np.where(d>1e-4)[0][-1]] for d in datYList])

    colorList = ['r', 'b', 'r', 'b', 'orange', 'purple']
    labelList = ['Data: H0', 'Data: H1']
    for i in range(len(StatList)):
        ax.plot(DatX, datYGen(StatList[i]), color=colorList[i], linestyle='-', label=labelList[i])
    
    x = np.linspace(xMin,xMax,200)
    labelList2 = ['Prediction: H0','Prediction: H1','Asimov: H0','Asimov: H1']
    for i in range(len(mu)):
        if i>1:
            line='-.'
        else:
            line=':'
        y1 = stats.norm.pdf(x,loc=mu[i],scale=sig[i])
        ax.plot(x, y1, color=colorList[i], linestyle=line, label=labelList2[i])
    
    lfs=25
    ax.set_xlabel(r"$q$",fontsize=lfs)
    ax.set_ylabel(r"Probability",fontsize=lfs)
    ax.legend(fontsize=17,frameon=False,loc=(0.65,0.67))
    #ax.set_yscale('log')
    ax.set(xlim=xrange,ylim=(0, ymax))
    ax.tick_params(which='major',direction='in',width=2,length=13,pad=10,labelsize=20)
    ax.tick_params(which='minor',direction='in',width=1,length=10)
    return 0
```
