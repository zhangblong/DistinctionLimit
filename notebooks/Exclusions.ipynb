```python
ne = 10 # number of energy bins (anything >50 is accurate enough)
nm = 200 # number of mass points
E_th = 1e-3*1e-6 # Threshold
E_max = 1e2*1e-6 # Max recoil energy
Nuc = Xe131
```

```python
myPath = os.path.join(os.path.abspath(os.path.pardir),'data','detection')
Xenon1TEffDat = np.transpose(np.loadtxt(os.path.join(myPath,'Xenon8BEfficiency.csv'),delimiter=','))
Xenon1TEffFunc = lambda x: np.interp(x, Xenon1TEffDat[0], Xenon1TEffDat[1])

PandaX4TEffDat = np.transpose(np.loadtxt(os.path.join(myPath,'PandaX4TEfficiency.txt')))
PandaX4TEffFunc = lambda x: np.interp(x, PandaX4TEffDat[0], PandaX4TEffDat[1])
```

```python
from WIMPFuncs import C_SI, MeanInverseSpeed_SHM, WIMPDiffRate
from LabFuncs import FormFactorHelm

def DMvDiffRatesGen(mDM, E_th, E_max, ne, Nuc, NuclearEnhancementFactor,v_lab=245.6):
    return WIMPDiffRate(E_th*1.0e6,E_max*1.0e6,ne,np.array([mDM]),Nuc,NuclearEnhancementFactor,FormFactorHelm,\
                                               lambda v_min: MeanInverseSpeed_SHM(v_min,sig_v=167.0,v_esc=533.0,v_lab=v_lab))

RDMN = DMvDiffRatesGen(6., E_th, E_max, ne, Nuc, C_SI)
```

```python
def DMLimitParaGenPadnaX4T(E_th,E_max,EffFunc,DMMassList):
    nuGen = NuRates(E_th,E_max,ne,Nuc)
    temp = nuGen.vDifferentialRatesSMCheck()
    x, yList = temp[1], temp[0]
    index8B = np.where(nuGen._Names == '8B')[0][0]
    n8BSM = yList[index8B]
    nNuSM = np.trapz(n8BSM*EffFunc(x),x)
    print(nNuSM*0.48)
    nLimit = 9/5.46*nNuSM
    print(nLimit*0.48)
    
    DMMassList=DMvRatesGen(DMMassList, E_th, E_max, ne, Nuc, C_SI)[0]
    nDMGen = lambda dat: np.trapz(dat[1]*EffFunc(dat[0]),dat[0])
    nDMList = np.array([nDMGen(DMvDiffRatesGen(m, E_th, E_max, ne, Nuc, C_SI)) for m in DMMassList])
    SMLimit = (nLimit-nNuSM)/nDMList*1e-45
    
    ZeroLimit = (nLimit-0)/nDMList*1e-45
    
    modelPara = [np.log10(2e-5),np.log10(30e-3),1,1,1]
    temp1 = nuGen.vDifferentialRatesFICheck(dRdEU1FI, \
              modelPara)
    x1, yList1 = temp1[1], temp1[0]
    n8BU1 = yList1[index8B]
    nNuU1 = np.trapz(n8BU1*EffFunc(x),x)
    U1Limit1 = (nLimit-nNuU1)/nDMList*1e-45
    
    modelPara = [np.log10(2.5e-5),np.log10(20e-3),1/3,1/3,-1]
    temp1 = nuGen.vDifferentialRatesFICheck(dRdEU1FI, \
              modelPara)
    x1, yList1 = temp1[1], temp1[0]
    n8BU1 = yList1[index8B]
    nNuU1 = np.trapz(n8BU1*EffFunc(x),x)
    U1Limit2 = (nLimit-nNuU1)/nDMList*1e-45
    return [SMLimit, U1Limit1, U1Limit2, ZeroLimit], DMMassList
```
